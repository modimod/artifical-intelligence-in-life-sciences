```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
folder = '1534167389048'
```

```python
class_auc = np.genfromtxt('saves/{}/class_auc.csv'.format(folder),delimiter=' ')
macro_auc = np.genfromtxt('saves/{}/macro_auc.csv'.format(folder))
```

```python
plt.figure(figsize=(10,10))
for i in range(class_auc.shape[1]):
    plt.plot(class_auc[:,i],label='Task {}'.format(i))
plt.plot(macro_auc,'bo',label='macro')
plt.xlabel('epochs')
plt.ylabel('auc')
plt.title('AUC for Classes and Macro')
plt.legend()
plt.grid(True)
```
