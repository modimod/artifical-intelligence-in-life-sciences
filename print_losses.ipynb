```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
import os
```

```python
folder = '1534167389048'
```

```python
losses = np.loadtxt(os.path.join('saves',folder,'losses_train.csv'))
losses_test = np.loadtxt(os.path.join('saves',folder,'losses_test.csv'))
accs_test = np.loadtxt(os.path.join('saves',folder,'accs_test.csv'))
```

```python
fig, (ax1,ax2) = plt.subplots(2, figsize=(15,15), sharex=True)

ax1.plot(losses,label='loss train')
ax1.plot(losses_test,label='loss validation')
ax2.plot(accs_test, label='Validation')

ax1.set_xlabel('epochs')
ax1.legend()
ax2.legend()
ax1.grid(True)
ax2.grid(True)
ax1.set_ylabel('loss')
ax2.set_ylabel('accuracy')

ax1.set_title('Losses')
ax2.set_title('Accuracy Validation')#, Best: {:.2%}'.format(accs_test[-1]))



plt.tight_layout()
```
